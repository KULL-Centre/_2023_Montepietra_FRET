```python
import MDAnalysis
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import itertools
import warnings
import pickle

warnings.filterwarnings('ignore')
```

```python
results_pp11_ws = np.load('results_pp11_ws.npy', allow_pickle=True)
results_pp11_ws_df = pd.DataFrame.from_records(results_pp11_ws)
```

```python
pp11_df = pd.DataFrame({'E': [0.736, 0.880, 0.993, 0.870, 0.83,
                              0.732, 0.876, 0.993, 0.867, 0.83],
                        'Regime': np.tile(['Static', 'Dynamic', 'Dynamic+', 'Average', 'MD'], 2),
                        'Calculation': np.repeat(['Reweighting', 'No Reweighting'], 5)
                       })
```

```python
sns.set_style('whitegrid')

g = sns.catplot(data=pp11_df, 
                x='Regime', y='E', col='Calculation',
                kind="bar", legend=True)

ax = g.axes

g.fig.set_size_inches(5.2, 3.7)

ax[0, 0].axhline(0.88, color='red', linestyle = '--')
ax[0, 1].axhline(0.88, color='red', linestyle = '--')

ax[0, 0].set_ylim(0, 1.1)
ax[0, 0].set_yticks(np.arange(0, 1.1, 0.1))
        
for col in [0, 1]:
    
    for label in (ax[0, col].get_xticklabels() + ax[0, col].get_yticklabels()):
        label.set_fontsize(8)
        
    for tick in ax[0, col].get_xticklabels():
        tick.set_rotation(30)

ax[0, 0].set_xlabel('Regime', fontsize=8)
ax[0, 1].set_xlabel('Regime', fontsize=8)
ax[0, 0].set_ylabel('FRET efficiency', fontsize=8)

ax[0, 0].set_title(None)
ax[0, 1].set_title(None)

# ax[0, 0].text(4.3, 1.2, 'pp11')
ax[0, 0].text(-0.3, 1.04, 'a', fontsize=12)
ax[0, 1].text(-0.3, 1.04, 'b', fontsize=12)

# plt.tight_layout()

# g.savefig('pp11_cutoff20_reweighting.eps')
```
